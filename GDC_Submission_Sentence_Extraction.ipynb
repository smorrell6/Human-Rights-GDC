```python
directory_path = '/content/drive/MyDrive/Northeastern/Thesis/Paper 3: Human Rights & The GDC/GDC Submissions'
count = 0
for filename in os.listdir(directory_path):
  filepath = os.path.join(directory_path, filename)
  if os.path.isfile(filepath):
    print(filename)
    print('\n')
    count += 1
print(count)
```

```python
#assisted by Colab AI
def extract_sentences(directory):
  total_df = pd.DataFrame(['file','sentence'])
  count = 0
  for filename in os.listdir(directory_path):
    filepath = os.path.join(directory_path, filename)
    print(filepath)
    count += 1
    reader = PdfReader(filepath) #reading in the PDF
    for page in reader.pages:
        list_page = page.extract_text().split(".")
        for sent in list_page:
          if len(sent.split()) > 1:
            #from Google AI
            new_row = {'file': [filename], 'sentence': [sent]} #https://www.geeksforgeeks.org/different-ways-to-create-pandas-dataframe/
            new_row_df = pd.DataFrame(new_row)
            total_df = pd.concat([total_df, new_row_df], ignore_index=True)
  print(count)
  return total_df
```

```python
sentence_df = extract_sentences('/content/drive/MyDrive/Northeastern/Thesis/Paper 3: Human Rights & The GDC/GDC Submissions')
```
