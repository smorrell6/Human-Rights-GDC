```python
import re
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import pyLDAvis
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk

# Download required NLTK data
nltk.download('punkt')
nltk.download('stopwords')
# Download the 'punkt_tab' resource
nltk.download('punkt_tab') # This line is added to download the missing resource
```

```python
# prompt: preprocess text in CSV rows

def preprocess_text(text):
    # Lowercase the text
    text = text.lower()
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenize the text
    words = word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [w for w in words if not w in stop_words]
    # Join the words back into a string
    text = " ".join(words)
    return text

# Sample DataFrame (replace with your actual data)
data = {'text_column': ["This is a sample sentence.", "Another sentence with some words."]}
df = pd.DataFrame(data)

# Apply the preprocessing function to the 'text_column'
df['processed_text'] = df['text_column'].apply(preprocess_text)

# Print the updated DataFrame
df
```

```python
# Load sentences
df_gdc = pd.read_csv('/content/gdc_submission_sentences_4Dec24 - english_gdc_submission_sentences_4Dec24 (2).csv')
```

```python
df_gdc
```

```python
df_gdc['processed_text'] = df_gdc['sentence'].apply(preprocess_text)
```

```python
df_gdc.head()
```

```python
# prompt: covert df_gdc rows to documents for topic modeling

# Assuming df_gdc is already loaded and preprocessed as shown in the provided code.

documents = df_gdc['processed_text'].tolist()
# Now 'documents' is a list where each element is a preprocessed document ready for topic modeling.

# Example usage with LDA (you can adapt this part for other topic modeling methods):
vectorizer = CountVectorizer(max_df=0.95, min_df=2, max_features=1000, stop_words='english')
dtm = vectorizer.fit_transform(documents)


```

```python
lda = LatentDirichletAllocation(n_components=11, random_state=42) # Adjust n_components as needed
lda.fit(dtm)
```

```python
lda_vis = pyLDAvis.prepare(topic_term_dists=lda.components_,
                           doc_topic_dists=lda.transform(dtm),
                           doc_lengths=[len(d.split()) for d in documents], # Added: Calculate document lengths
                           vocab=vectorizer.get_feature_names_out(), # Added: Get vocabulary from vectorizer
                           term_frequency=dtm.sum(axis=0).tolist()[0]) # Added: Get term frequency from doc_term_matrix

pyLDAvis.save_html(lda_vis, 'topic_model_visualization.html')
```

```python
pyLDAvis.display(lda_vis)
```

```python
# prompt: make a table showing the top words for each topic

def display_top_words_per_topic(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic %d:" % (topic_idx))
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))

# Assuming 'lda' is your fitted LDA model and 'vectorizer' is your CountVectorizer
no_top_words = 20
display_top_words_per_topic(lda, vectorizer.get_feature_names_out(), no_top_words)


# Create a table for better visualization

topic_words = []
for topic_idx, topic in enumerate(lda.components_):
    top_words = [vectorizer.get_feature_names_out()[i] for i in topic.argsort()[:-no_top_words - 1:-1]]
    topic_words.append(top_words)

topics_df = pd.DataFrame(topic_words).transpose()
topics_df.columns = [f'Topic {i}' for i in range(len(topic_words))]
topics_df
```

```python
# prompt: assign the most prominent topic for each row in df_gdc[processed_text] to a new column called topic

import numpy as np

# Get the dominant topic for each document
dominant_topics = np.argmax(lda.transform(dtm), axis=1)

# Add the dominant topic as a new column to the DataFrame
df_gdc['topic_1'] = dominant_topics
```

```python
# prompt: get top three topics for each sentence

def get_top_three_topics(lda_model, dtm, vectorizer, document_index):
    """
    Gets the top three topics for a specific document.

    Args:
        lda_model: The fitted LDA model.
        dtm: The document-term matrix.
        vectorizer: The CountVectorizer used to create the DTM.
        document_index: The index of the document in the DataFrame.

    Returns:
        A list of strings representing the top three topics for the document.
    """
    topic_probabilities = lda_model.transform(dtm[document_index])
    top_three_topic_indices = np.argsort(topic_probabilities[0])[-3:][::-1]
    top_three_topics = []
    for topic_index in top_three_topic_indices:
      top_words = [vectorizer.get_feature_names_out()[i] for i in lda_model.components_[topic_index].argsort()[:-6:-1]]
      top_three_topics.append(" ".join(top_words))
    return top_three_topics


# Example usage:
# Assuming 'lda', 'dtm', 'vectorizer', and 'df_gdc' are already defined as in your original code

# Get top 3 topics for sentence at index 0
top_topics = get_top_three_topics(lda, dtm, vectorizer, 0)
print(f"Top 3 topics for sentence 0: {top_topics}")


# To get top topics for all sentences
df_gdc['top_3_topics'] = [get_top_three_topics(lda, dtm, vectorizer, i) for i in range(len(df_gdc))]
print(df_gdc[['sentence', 'top_3_topics']].head())
```
