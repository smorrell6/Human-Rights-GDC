```python
import pandas as pd
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/TechnicalCommunityActors_QCA_18March25 - Binary Categories.csv')
```

```python
df_cats = df.columns[3:]
```

```python
# prompt: remove Binary frpm the end of each column name

new_column_names = [col.replace('Binary', '').strip() for col in df_cats]
df.rename(columns=dict(zip(df_cats, new_column_names)), inplace=True)
df.head()
```

```python
df_cats = df.columns[3:]
```

```python
df_cats
```

```python
# prompt: make a co-occurrence network using the binary categories


# Assuming 'df' is your DataFrame with binary categories

# Create an empty graph
graph = nx.MultiGraph()

```

```python
# Iterate through rows and create edges based on co-occurrence
for index in df.index:  # Use df.index to iterate over DataFrame rows
  row = df.loc[index, df_cats]  # Get the specific row's data for the relevant columns
  for i in range(len(df_cats)):  # Iterate over the binary category columns
    if row.iloc[i] == 1:  # Check the value using iloc for integer-based indexing
      for j in range(i + 1, len(df_cats)):
        if row.iloc[j] == 1:
          graph.add_edge(df_cats[i], df_cats[j])  # Use df_cats for correct column names

# Visualize the network
nx.draw(graph, with_labels=True, node_size=1500, font_size=10)
plt.show()
```

```python
# prompt: print all graph edges

# Print the edges with their counts, from Google search AI
for u,v, data in graph.edges(data=True):
    print(f"Edge: {u,v}, Count: {graph.number_of_edges(u,v)}")
```

```python
# prompt: graph network so that edges are weighted by their count

# Create an empty graph
graph = nx.Graph()

# Iterate through rows and create edges based on co-occurrence
for index in df.index:
  row = df.loc[index, df_cats]
  for i in range(len(df_cats)):
    if row.iloc[i] == 1:
      for j in range(i + 1, len(df_cats)):
        if row.iloc[j] == 1:
          if graph.has_edge(df_cats[i], df_cats[j]):
            graph[df_cats[i]][df_cats[j]]['weight'] += 1
          else:
            graph.add_edge(df_cats[i], df_cats[j], weight=1)

# Visualize the network with edge weights
pos = nx.spring_layout(graph)
nx.draw(graph, pos, with_labels=True, node_size=1500, font_size=10)
edge_labels = nx.get_edge_attributes(graph, 'weight')
nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels)
plt.show()

# Print the edges with their weights
for u, v, data in graph.edges(data=True):
  print(f"Edge: {u, v}, Weight: {data['weight']}")
```

```python
# prompt: color the graph edges by weight
# prompt: size nodes by degree

# Visualize the network with edge weights and color by weight
fig, ax = plt.subplots(figsize = (15,10)) # Create a figure and an axes object
pos = nx.spring_layout(graph, k=2, iterations=20) #https://stackoverflow.com/questions/14283341/how-to-increase-node-spacing-for-networkx-spring-layout
edges, weights = zip(*nx.get_edge_attributes(graph, 'weight').items())

# Define a colormap for the edges based on weight
# Use a diverging colormap to highlight both low and high weights
cmap = plt.cm.get_cmap('Blues')

# Normalize edge weights for color mapping
norm = plt.Normalize((min(weights) - 20), max(weights))

# Draw edges with colors based on weights
nx.draw(graph, pos, with_labels=True, node_size=1500, font_size=10, edge_color=[cmap(norm(w)) for w in weights], ax=ax, node_color = 'lightgrey') # Specify the axes to draw on

# Draw edge labels
edge_labels = nx.get_edge_attributes(graph, 'weight')
nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels, ax=ax)  # Specify the axes to draw on

# Now, matplotlib knows where to place the colorbar
plt.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap), label='Edge Weight', ax=ax) # Specify the axes to steal space from
plt.show()

# Print the edges with their weights
for u, v, data in graph.edges(data=True):
  print(f"Edge: {u, v}, Weight: {data['weight']}")
```
